```python
import pickle

c = pickle.load(open('counter.pkl', 'rb'))
wordcounts = np.array(list(c.values()))
wordcounts.sort()
```

```python
vocabsize = [len(wordcounts[wordcounts >= n]) for n in range(1, 101)]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import itertools as it

sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
plt.plot(range(1, 101), vocabsize)
plt.xticks([1, 2, 3, 4, 5, 7] + list(range(10, 101, 5)))
plt.yticks([100_000, 500_000, 1_000_000, 1_500_000, 2_000_000, 3_000_000, 4_500_000])
plt.xlabel('Minimum number of occurrences')
plt.ylabel('Number of words')
plt.show()
```
